```python
records = [
    {"name": "John Doe", "email": "john@example.com", "sales": 150},
 {"name": "Alice", "email": "alice@", "sales": -20},
    {"name": "", "email": "bob@example.com", "sales": 90},
]

def check_record(record):
    if record["name"] == "":
        return False
    elif "@" not in record["email"]:
        return False
    elif record["sales"] <= 0:
        return False
    return True
clean_data = []
bad_data = []
for record in records:
    if check_record(record):
        clean_data.append(record)
    else:
        bad_data.append(record)
print("Clean Data:", clean_data)
print("Bad Data:", bad_data)

```
